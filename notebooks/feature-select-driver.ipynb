```python
import pandas as pd
import sys
sys.path.insert(0,'../src')
from feature_selection_driver import FeatureSelectionDriver
from driver import Driver
from etl.zoo_etl import ZooEtl
from etl.iris_etl import IrisEtl
```

```python
feature_dfs = []
```

```python
def run_test(data_set,model_params,fimp_type,acc_percent_threshold=.9):
    if (model_params['model_type']=='log_reg')&(fimp_type=='shap'):
        return None
    driver = FeatureSelectionDriver(data_set, model_params,fimp_type)
    feature_df = driver.run(acc_percent_threshold, verbose=False)
    
    feature_df['data_set']=data_set
    feature_df['model_type']=model_params['model_type']
    feature_df['fimp_type']=fimp_type
    
    return feature_df
```

```python
data_sets = ['iris','zoo','credit']
```

```python
model_params_arr = [
    {'model_type':'log_reg', 'max_iter':100},
    {'model_type':'rf_class', 'n_estimators':100, 'max_depth':10},
    {'model_type':'nn', 'num_hidden_layers':4, 'nodes_per_layer':64, 'activation':'relu', 
     'optimizer':'adam', 'loss':'crossentropy', 'metrics':['accuracy'], 'num_epochs':20}
]
```

```python
fimp_types = ['shap', 'perm']
```

```python
verbose = True
for data_set in data_sets:
    for model_params in model_params_arr:
        for fimp_type in fimp_types:
            if (fimp_type=='shap')&(model_params['model_type']=='log_reg'):
                continue
            if verbose:
                print('Data:{}, Model:{}, Fimp:{}'.format(data_set,model_params['model_type'],fimp_type))
            feature_df = run_test(data_set,model_params,fimp_type)
            feature_dfs.append(feature_df)
```
